```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
# Load temperature data 
df = pd.read_csv('temperature_data.csv')
df.Data_Value = df.Data_Value/10
df.sort_values(['Date','Element'],inplace=True)
df.head()
```

```python
# Process Maximum Daily Temperatures

# Step 1: Prepare Max temp data
tmax = df.groupby(['Date','Element']).max()
tmax = pd.DataFrame(pd.pivot_table(tmax, index=[tmax.index.get_level_values(0)], columns=[tmax.index.get_level_values(1)], values=['Data_Value'])['Data_Value']['TMAX'])
tmax.index = pd.to_datetime(tmax.index)
tmax['Year'] = tmax.index.year
tmax['Month'] = tmax.index.month
tmax['Day'] = tmax.index.day
tmonth = tmax['Month']

# Step 2: Remove Leap days and year 2015
tmax2015 = tmax[(tmax.Year == 2015)]
tmax2015 = tmax2015.drop('Year', axis=1).set_index(['Month','Day'])
tmax = tmax[~((tmax.Month == 2) & (tmax.Day == 29)) & (tmax.Year != 2015)]
tmax.reset_index(inplace=True)
tmax.drop(['Date','Year'], axis=1, inplace=True)
tmax = tmax.groupby(['Month','Day']).agg(max)
tmax.reset_index(inplace=True)
```

```python
# Process Minimum Daily Temperatures

# Step 1: Prepare Min temp data
tmin = df.groupby(['Date','Element']).min()
tmin = pd.DataFrame(pd.pivot_table(tmin, index=[tmin.index.get_level_values(0)], columns=[tmin.index.get_level_values(1)], values=['Data_Value'])['Data_Value']['TMIN'])
tmin.index = pd.to_datetime(tmin.index)
tmin['Year'] = tmin.index.year
tmin['Month'] = tmin.index.month
tmin['Day'] = tmin.index.day

# Step 2: Remove Leap days and year 2015
tmin2015 = tmin[(tmin.Year == 2015)]
tmin2015 = tmin2015.drop('Year', axis=1).set_index(['Month','Day'])
tmin = tmin[~((tmin.Month == 2) & (tmin.Day == 29)) & (tmin.Year != 2015)]
tmin.reset_index(inplace=True)
tmin.drop(['Date','Year'], axis=1, inplace=True)
tmin = tmin.groupby(['Month','Day']).agg(min)
tmin.reset_index(inplace=True)
```

```python
# Processing points for the Year 2015

tmax2015.reset_index(inplace=True)
tmax2015 = tmax2015[tmax2015.TMAX>tmax.TMAX]
tmin2015.reset_index(inplace=True)
tmin2015 = tmin2015[tmin2015.TMIN<tmin.TMIN]
```

```python
# Plotting the Line Graph (2005-14) and Scatter Plot (2015)

plt.figure(figsize=(16,8))
plt.plot(tmax.reset_index()['TMAX'], color='red', alpha=0.6);
plt.plot(tmin.reset_index()['TMIN'], color='blue', alpha=0.6);
plt.scatter(x=tmax2015.index, y=tmax2015.TMAX, color='maroon');
plt.scatter(x=tmin2015.index, y=tmin2015.TMIN, color='blue');
plt.xticks(np.linspace(15,15 + 30*11 , num = 12),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

plt.gca().fill_between(range(len(tmax)),
                       tmax['TMAX'], tmin['TMIN'],
                       facecolor='grey', alpha=0.15)

plt.ylabel('Temperature ($^o$C)')
plt.title('Daily Extreme Temperatures (2005-2014), with Record Breaking Temperature Points (2015)')
# add a legend with legend entries
plt.legend(['Maximum Temperature (2005-14)', 
            'Minimum Temperature (2005-14)', 
            'Maximum Record Temperature (2015)',
            'Minimum Record Temperature (2015)']);
```
